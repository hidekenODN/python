### Chapter2：基本文法

```python
#print関数の改行
print("日本国民は", end="\r\n")      #CRLF改行
print("正当に選挙された", end="\n")  #LF改行
print("国会の代表者を通じて行動する")
```

```python
#formatで植え込む
#書式："{key1}, {key2}".format(key1=value1, key2=value2)

print("私の名前は{name}です。".format(name="田中太郎"))
```

### Chapter3：Listや関数について

```python
#listを用いたfor文
list = [1,2,3,4,5]
sum = 0
for i in list:
    sum += i
print(sum)    
```

```python
#enumerate構文：リストにインデックスを付けたいとき
list = ["Apple", "Banana", "Orange"]
for i,v in enumerate(list):
    print(i,v)
print("\r\n")
```

```python
#活用例：パラメータを段階的に変化させたいとき
#下記例：定義域xim=[-10,10]内部での関数fの値域ylimを、関数のパラメータごとに計算したい
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

f = lambda a, x: np.cos(a * x)
param = np.array([0.1, 1.0, 2.0])
xlim = np.arange(-10, 10+1, 0.1)
ylim = np.zeros([param.shape[0], x.shape[0]], dtype=float)

for i,j in enumerate(param):
    ylim[i,:] = f(param[j],x)

fig = plt.figure()
ax = plt.subplot(1,1,1)
ax.set_xlabel("x")
ax.set_ylabel("cos(x)")
ax.set_xlim([-10,10])
ax.plot(xlim, ylim[0,:])
ax.plot(xlim, ylim[1,:])
ax.plot(xlim, ylim[2,:])
```
