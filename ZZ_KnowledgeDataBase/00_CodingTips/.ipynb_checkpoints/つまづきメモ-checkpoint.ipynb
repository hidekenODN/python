### Pythonデータ構造の生成方法まとめ

- list：`[x1, x2, x3,....]`
- dict：`{key1:value1, key2:value2, ...}`
- numpy.ndarray
    - `np.array(list)`
- pandas.Series
    - `pd.Series(list, index=list)`
    - `pd.Series(tuple, index=list)`
    - `pd.Series(dict)`
    - `pd.Series(np.ndarray, index=list)`
- pandas.DataFrame
    - `pd.DataFrame(list, index=list, columns=list)`
    - `pd.DataFrame(Series,  columns=list)`
    - `pd.DataFrame(dict)`
    - `pd.DataFrame(np.ndarray, columns=list)`

#### サンプルコード：

```python
# list型：
A = [1,2,3,4]
A
```

```python
#ndarray型：
import numpy as np
B = np.array([1,2,3])
B
```

```python
#Series型：
from pandas import Series
import numpy as np
C1 = Series([1,2,3,4], index=['A', 'B', 'C', 'D'])
C2 = Series((1,2,3), index=['A', 'B', 'C'])
C3 = Series({1:'A', 2:'B'})
C4 = Series(np.array([1,2,3]), index=['P', 'Q', 'R'])
print(C1)
print(C2)
print(C3)
print(C4)
```

```python
#DataFrame型：
from pandas import DataFrame
D1 = DataFrame([[1,2,3],[4,5,6],[7,8,9]], index=['i1', 'i2', 'i3'], columns=['c1', 'c2', 'c3'])
D2 = DataFrame(C1, columns=['c1'])
D3 = DataFrame({'key1':[1,2,3], 'key2':[4,5,6]})
D4 = DataFrame(np.array([[1,2,3],[4,5,6]]), columns=['P', 'Q', 'R'])
print(D1)
print(D2)
print(D3)
print(D4)
```

### range, np.arangeの違い

- range：range型
    - 書き方：`range(start, stop[, step])`
    - 用途：数のイミュータブルなシーケンスを表す。
        - 特定の回数のループ：`for i in range(N)`
        - 連番や等差数列を持ったlist型：`[x for in range(N)]`
        - 連番や等差数列を持ったnp.array型：`np.array(range(N))`
- np.arange：np.ndarray型
    - 書き方：`numpy.arange([start, ]stop, [step, ]dtype = None)`
    - 用途：rangeより高速に連番や等差数列を持ったnp.array型を生成する
        - 連番や等差数列を持ったnp.array型：`np.arange(N)`
- linspace

### nd.arrayの転置

線形代数でのベクトル$\boldsymbol{x} = [1,2,3]$は1次元テンソルであり、転置$\boldsymbol{x}^T $が可能。  
これに対応する量をPython上で実装する際は`x=np.array([1,2,3])`を連想する。
しかしこの変数は1次元配列`x.shape=(3,)`であり、このままでは転置`.T`が出来ない。その代わり、以下の方法をとる必要がある。

- 元から2次元配列として定義する：`np.array([[1,2,3]]).T`
- reshapeする（この場合、要素数を知る必要があり、コードが煩雑に）：`x.reshape(3,1)`
- np.newaxisで次元を拡張する：`a[:,np.newaxis]`, `a[np.newaxis,:].T`

### matplotlibを用いた3次元プロット

参照：https://qiita.com/kazetof/items/c0204f197d394458022a  
参照：https://matplotlib.org/tutorials/toolkits/mplot3d.html#sphx-glr-tutorials-toolkits-mplot3d-py

```python
import numpy as np
from scipy.stats import multivariate_normal

#for plotting
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

m = 2 #dimension
mean = np.zeros(m)
sigma = np.eye(m)
```

#### Surface Plot
Surface Plot（日本語でいうと表面プロット？）を試す．
注意点としては，plot_surface関数に渡すデータは，2次元配列になっているところ．

```python
N = 1000
x1 = np.linspace(-5, 5, N)
x2 = np.linspace(-5, 5, N)
X1, X2 = np.meshgrid(x1, x2)

# 領域全体の各座標に一度に演算を行うため、一度np.ravel(*)で*を1次元化し、np.c_でhstackでまとめる
# X.shape=(1000000, 2)になる
X = np.c_[np.ravel(X1), np.ravel(X2)] 

# Y_plot.shape=(1000000, 2)
Y_plot = multivariate_normal.pdf(x=X, mean=mean, cov=sigma)

# Y_plot.shape=(1000, 1000)に変更
Y_plot = Y_plot.reshape(X1.shape)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(X1, X2, Y_plot, cmap='bwr', linewidth=0)
fig.colorbar(surf)
ax.set_title("Surface Plot")
fig.show()
```

```python
Y_plot.shape
```

#### Contour Plot
Contour Plot（等高線プロット）もsurface plotと同じようにできる．

```python
N = 1000
x1 = np.linspace(-5, 5, N)
x2 = np.linspace(-5, 5, N)

X1, X2 = np.meshgrid(x1, x2)
X = np.c_[np.ravel(X1), np.ravel(X2)]
Y_plot = multivariate_normal.pdf(x=X, mean=mean, cov=sigma)
Y_plot = Y_plot.reshape(X1.shape)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.contour(X1, X2, Y_plot)
ax.set_title("Contour Plot")
fig.show()
```

#### Scatter Plot
今までとは異なり，scatter plot（散布図）に渡すデータは1次元配列である．

```python
N = 100
x1 = np.linspace(-5, 5, N)
x2 = np.linspace(-5, 5, N)

X1, X2 = np.meshgrid(x1, x2)
X_plot = np.c_[np.ravel(X1), np.ravel(X2)]

y = multivariate_normal.pdf(X_plot, mean=mean, cov=sigma)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter3D(np.ravel(X1), np.ravel(X2), y)
ax.set_title("Scatter Plot")
plt.show()
```
